```python
import xarray as xr
import matplotlib.pylab as plt
import numpy as np
```

```python
fn1="/gws/nopw/j04/class_vol2/senemo/RUNS2024r01/GS1p2_full/output/SENEMO_1y_19810101_19811231_grid_T_2D.nc"
f1=xr.open_dataset(fn1)
fn2="/gws/nopw/j04/class_vol2/senemo/FINAL_PHYSICS_RUNS/GS1p2_full/output_old/SENEMO_1y_19810101_19811231_grid_T_2D.nc"
f2=xr.open_dataset(fn2)

```

```python
M2_1=np.sqrt(f1.M2x.values[:,:]**2+f1.M2y.values[:,:]**2)
M2_2=np.sqrt(f2.M2x.values[:,:]**2+f2.M2y.values[:,:]**2)
```

```python
plt.pcolormesh(M2_1,vmax=1);plt.colorbar(orientation='vertical')
```

```python
plt.pcolormesh(M2_2,vmax=1);plt.colorbar(orientation='vertical')
```

```python
plt.pcolormesh(M2_2-M2_1,vmin=-0.1,vmax=0.1);plt.colorbar(orientation='vertical')
```

```python
plt.pcolormesh(M2_2/M2_1-1,vmin=-0.5,vmax=0.5);plt.colorbar(orientation='vertical')
```
